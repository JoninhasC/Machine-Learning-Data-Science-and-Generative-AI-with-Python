```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 50.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```

```python
def plotar_distribuicoes_normais(medias, desvios_padroes, num_amostras=10000, bins=50):
    plt.figure(figsize=(15, 10))
    
    for i, media in enumerate(medias):
        for j, desvio_padrao in enumerate(desvios_padroes):
            # Gerar dados
            dados = np.random.normal(media, desvio_padrao, num_amostras)
            
            # Plotar histograma
            plt.subplot(len(medias), len(desvios_padroes), i * len(desvios_padroes) + j + 1)
            plt.hist(dados, bins=bins, density=True, alpha=0.6, color='g', edgecolor='black')
            plt.title(f'Média = {media}, Desvio Padrão = {desvio_padrao}')
            plt.xlabel('Valor')
            plt.ylabel('Densidade')
    
    plt.tight_layout()
    plt.show()
```

```python
# Parâmetros para experimentar
medias = [0, 50, 100]        # Diferentes médias
desvios_padroes = [10, 25, 50]  # Diferentes desvios padrões

# Plotar as distribuições
plotar_distribuicoes_normais(medias, desvios_padroes)
```
